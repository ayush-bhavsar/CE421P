```python
# Import necessary libraries
import numpy as np
from sklearn.preprocessing import LabelEncoder

# Sample categorical data (each value represents a category)
data = np.array(['Red', 'Blue', 'Green', 'Red', 'Green'])

# Create a LabelEncoder object
encoder = LabelEncoder()

# Fit the encoder to the data and transform it
label_encoded_data = encoder.fit_transform(data)

# Print the original and label encoded data
print("Original Data:")
print(data)
print("\nLabel Encoded Data:")
print(label_encoded_data)
```
