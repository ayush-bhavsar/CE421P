```python
# Import pandas
import pandas as pd

# Create a dataset.
data = {
'Name': ['Amit', 'Neha', 'Rahul', 'Neha', 'Sita', 'Rohan', 'Priya', 'Aman'],
'Age': [20, 21, None, 21, 22, 23, None, 24],
'Marks': [85, 90, 88, 90, None, 76, 92, None],
'City': ['Delhi', 'Mumbai', 'Delhi', 'Mumbai', 'Pune', 'Delhi', 'Pune', 'Mumbai']
}

# Load dataset into DataFrame
df = pd.DataFrame(data)
print(df)

# Display dataset
print("Head:")
print(df.head())
print("\nTail:")
print(df.tail())

# Dataset information
print("\nInfo:")
print(df.info())
print("\nShape (rows, columns):")
print(df.shape)
print("\nDescribe:")
print(df.describe())

# Missing values
print("\nIs Null:")
print(df.isnull())
print("\nSum of missing values:")
print(df.isnull().sum())

# Mean, Median, Mode
print("\nMean:")
print(df.mean(numeric_only=True))
print("\nMedian:")
print(df.median(numeric_only=True))
print("\nMode:")
print(df.mode())

# Data slicing
print("\nSlicing rows 0 to 4:")
print(df[0:5])
print("\nSelects the last two rows of the DataFrame.")
print(df[-2:])
print("\nReturns the Name column as a Series.")
print(df['Name'])

# Using loc and iloc.
print("\nUsing loc:")
print(df.loc[0:4, ['Name', 'Marks']])
print("\nUsing iloc:")
print(df.iloc[0:5, 0:3])

# Fill missing values (correct way)
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Marks'] = df['Marks'].fillna(df['Marks'].mean())

# Duplicate handling
print("\nDuplicate rows:")
print(df.duplicated())
df.drop_duplicates(inplace=True)

# Final dataset
print("\nFinal Pre-Processed Dataset:")
print(df)
```
