```python
# Import libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Load built-in dataset
df = sns.load_dataset("tips")

# Display datset
print("\n Information.")
print(df.info)
print("\n Description")
print(df.describe())
print("\nFirst 5 rows:-")
print(df.head())

# BAR CHART: Average tip by day
sns.barplot(x='day', y='tip', data=df)
plt.title("Bar Chart: Average Tip by Day")
plt.show()

# LINE CHART: Total bill trend by party size
line_data = df.groupby('size')['total_bill'].mean().reset_index()
plt.plot(line_data['size'], line_data['total_bill'], marker='o', color='orange')
plt.xlabel("Party Size")
plt.ylabel("Average Total Bill")
plt.title("Line Chart: Total Bill by Party Size")
plt.show()

# PIE CHART: Smoker vs Non-Smoker
smoker_counts = df['smoker'].value_counts()
plt.pie(smoker_counts, labels=smoker_counts.index, autopct='%1.1f%%',
colors=['lightblue','lightgreen'])
plt.title("Pie Chart: Smoker vs Non-Smoker")
plt.show()

# HISTOGRAM: Total bill distribution
plt.hist(df['total_bill'], bins=10, color='skyblue', edgecolor='black')
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.title("Histogram: Total Bill Distribution")
plt.show()

# SCATTER PLOT: Total bill vs Tip
plt.scatter(df['total_bill'], df['tip'], color='green')
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.title("Scatter Plot: Total Bill vs Tip")
plt.show()

# AREA CHART: Total bill and tip over index
plt.stackplot(df.index, df['total_bill'], df['tip'], labels=['Total Bill','Tip'],
colors=['#FF9999','#99CCFF'])
plt.xlabel("Index")
plt.ylabel("Amount")
plt.title("Area Chart: Total Bill and Tip")
plt.legend()
plt.show()

# BOX PLOT: Total bill by day
sns.boxplot(x='day', y='total_bill', data=df, palette="pastel")
plt.title("Box Plot: Total Bill by Day")
plt.show()

# HEATMAP: Correlation of numeric columns
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Heatmap: Correlation of Numeric Columns")
plt.show()
```
