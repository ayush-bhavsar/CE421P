```python
# Import necessary libraries
import numpy as np
from sklearn.preprocessing import StandardScaler

# Sample data (each row represents a data point, each column represents a feature)
data = np.array([[50, 200], [30, 400], [20, 300], [80, 100]])

# Create a StandardScaler object to scale the data
scaler = StandardScaler()

# Fit the scaler to the data and transform it
standardized_data = scaler.fit_transform(data)

# Print the original and standardized data
print("Original Data:")
print(data)
print("\nStandardized Data:")
print(standardized_data)
```
