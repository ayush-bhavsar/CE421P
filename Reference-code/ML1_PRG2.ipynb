```python
# Import NumPy
import numpy as np

# FUNCTIONS FOR 1-D ARRAY
# Create 1-D arrays
print("\n Creating the 1-D Array.")
a = np.arange(1, 6) # arange
b = np.zeros(5) # zeros
c = np.ones(5) # ones
d = np.linspace(1, 10, 5) # linspace
print("Arange:", a)
print("Zeros:", b)
print("Ones:", c)
print("Linspace:", d)

# Indexing and Slicing
print("\nIndexing a[2]:", a[2])
print("Slicing a[1:4]:", a[1:4])

# STATISTICAL FUNCTIONS
print("\nThe Statistical Functions are:-")
print("Mean:", np.mean(a))
print("Median:", np.median(a))
print("Min:", np.min(a))
print("Max:", np.max(a))
print("Sum:", np.sum(a))
print("Product:", np.prod(a))
print("Variance:", np.var(a))
print("Standard Deviation:", np.std(a))

# Covariance (needs 2 arrays)
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])
print("Covariance:\n", np.cov(x, y))

# NUMERICAL OPERATIONS
print("\nThe Numerical Operations are:-")
print("Sorted array:", np.sort(a))
print("Power a^2:", np.power(a, 2))
print("Square root:", np.sqrt(a))
print("Absolute:", np.abs(-a))
print("Sin:", np.sin(a))
print("Cos:", np.cos(a))
print("Tan:", np.tan(a))
print("Log:", np.log(a))
print("Exponential:", np.exp(a))

# 1-D ARRAY OPERATORS
p = np.array([10, 20, 30])
q = np.array([2, 5, 10])

# Mathematical operators
print("\nThe Mathematical operator are:-")
print("Addition:", p + q)
print("Subtraction:", p - q)
print("Multiplication:", p * q)
print("Division:", p / q)
print("Floor Division:", p // q)
print("Modulus:", p % q)

# Relational operators
print("\nThe Relational operator are:-")
print("Greater than:", p > q)
print("Less than:", p < q)
print("Equal to:", p == q)

# 2-D ARRAY & OPERATORS
print("\n Creating a 2D Array.")
m = np.array([[1, 2], [3, 4]])
n = np.array([[5, 6], [7, 8]])
print("2-D Array m:\n", m)
print("2-D Array n:\n", n)

# Mathematical operators
print("\nThe Mathematical operator are:-")
print("Addition:\n", m + n)
print("Subtraction:\n", m - n)
print("Multiplication:\n", m * n)
print("Division:\n", m / n)

# Relational operators
print("\nThe Relational operator are:-")
print("Greater than:\n", m > n)
print("Equal to:\n", m == n)
```
