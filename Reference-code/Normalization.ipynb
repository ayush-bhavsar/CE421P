```python
# Import necessary libraries
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# Sample data (each row represents a data point, each column represents a feature)
data = np.array([[50, 200], [30, 400], [20, 300], [80, 100]])

# Create a MinMaxScaler object to scale the data between 0 and 1
scaler = MinMaxScaler(feature_range=(0, 1))

# Fit the scaler to the data and transform it
normalized_data = scaler.fit_transform(data)

# Print the original and normalized data
print("Original Data:")
print(data)
print("\nNormalized Data:")
print(normalized_data)
```
