```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# create dataset
data = pd.DataFrame({
    "x": np.arange(1, 11),
    "y": np.random.randint(10, 100, 10),
    "category": ["A","B","A","B","A","B","A","B","A","B"]
})

# line plot
plt.figure()
plt.plot(data["x"], data["y"])
plt.title("Line Plot")
plt.show()

# bar plot
plt.figure()
plt.bar(data["x"], data["y"])
plt.title("Bar Plot")
plt.show()

# scatter plot
plt.figure()
plt.scatter(data["x"], data["y"])
plt.title("Scatter Plot")
plt.show()

# histogram
plt.figure()
plt.hist(data["y"])
plt.title("Histogram")
plt.show()

# seaborn count plot
plt.figure()
sns.countplot(x="category", data=data)
plt.title("Count Plot")
plt.show()

# seaborn box plot
plt.figure()
sns.boxplot(x="category", y="y", data=data)
plt.title("Box Plot")
plt.show()

# seaborn heatmap
plt.figure()
corr = data.corr(numeric_only=True)
sns.heatmap(corr, annot=True)
plt.title("Heatmap")
plt.show()
```
