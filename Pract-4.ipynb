```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder

# create dataset
data = pd.DataFrame({
    "Name": ["A", "B", "C", "D", "E", "F", "G", "H"],
    "City": ["Delhi", "Mumbai", "Delhi", "Chennai", "Mumbai", "Delhi", "Chennai", "Mumbai"],
    "Department": ["HR", "IT", "Finance", "IT", "HR", "Finance", "IT", "HR"],
    "Age": [25, 30, 35, 28, 32, 40, 29, 31]
})

print("Original Data:\n")
print(data)

# check missing values
print("\nMissing Values:\n")
print(data.isnull().sum())

# fill missing values if any
data.fillna(method='ffill', inplace=True)

# separate categorical columns
categorical_cols = ["City", "Department"]

# apply one-hot encoding using pandas
encoded_df = pd.get_dummies(data, columns=categorical_cols)

print("\nOne-Hot Encoded Data using pandas:\n")
print(encoded_df)

# apply one-hot encoding using sklearn
encoder = OneHotEncoder(sparse_output=False)
encoded_array = encoder.fit_transform(data[categorical_cols])

# create dataframe from encoded array
encoded_cols = encoder.get_feature_names_out(categorical_cols)
encoded_sklearn_df = pd.DataFrame(encoded_array, columns=encoded_cols)

# combine with original numerical data
final_df = pd.concat([data.drop(columns=categorical_cols), encoded_sklearn_df], axis=1)

print("\nOne-Hot Encoded Data using sklearn:\n")
print(final_df)

# save to file
final_df.to_csv("one_hot_encoded_data.csv", index=False)

print("\nEncoding Completed Successfully")
```
