```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder

# create dataset
data = pd.DataFrame({
    "Name": ["A", "B", "C", "D", "E", "F", "G", "H"],
    "City": ["Delhi", "Mumbai", "Delhi", "Chennai", "Mumbai", "Delhi", "Chennai", "Mumbai"],
    "Department": ["HR", "IT", "Finance", "IT", "HR", "Finance", "IT", "HR"],
    "Gender": ["Male", "Female", "Female", "Male", "Male", "Female", "Male", "Female"],
    "Age": [25, 30, 35, 28, 32, 40, 29, 31]
})

print("Original Data:\n")
print(data)

# check missing values
print("\nMissing Values:\n")
print(data.isnull().sum())

# fill missing values if any
data.fillna(method='ffill', inplace=True)

# initialize label encoder
le = LabelEncoder()

# encode City column
data["City"] = le.fit_transform(data["City"])

# encode Department column
data["Department"] = le.fit_transform(data["Department"])

# encode Gender column
data["Gender"] = le.fit_transform(data["Gender"])

print("\nLabel Encoded Data:\n")
print(data)

# show mapping example
print("\nExample Mapping for Gender:")
for i, label in enumerate(le.classes_):
    print(label, "->", i)

# save to file
data.to_csv("label_encoded_data.csv", index=False)

print("\nEncoding Completed Successfully")
```
